```
import matplotlib.pyplot as plt
```

```
# PRIMEIRO GRAFICO

x = [1,2]
y = [2,3]

plt.plot(x,y)
plt.show()
```

```
#GRAFICO DE LINHAS 

x = [1,2,5]
y = [2,5,7]

# título
plt.title("Meu primeiro gráfico")

# eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.plot(x,y)
#plt.show()
```

```
# GRAFICO DE BARRAS

x = [1,2,3,4,5]
y = [2,5,7,1,0]

# título
plt.title("Grafico de barras")

# eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.bar(x,y) # grafico de barras
plt.show()
```

```
# GRAFICO DE BARRAS 2

x1 = [1,3,5,7,9]
y1 = [2,5,7,1,-1]

x2 = [2,4,6,8,10]
y2 = [12,15,17,11,10]

# título
plt.title("Grafico de barras")

# eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# grafico de barras
plt.bar(x1,y1, label = "Grupo 1") 
plt.bar(x2,y2, label = "Grupo 2")
plt.legend()

plt.show()
```

```
# GRAFICO DE DISPERSAO

x1 = [1,3,5,7,9]
y1 = [2,5,7,1,-1]


# título
plt.title("Grafico de Pontos")

# eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# grafico de dispersao
plt.scatter(x1,y1, label = "Pontos", color = "black") 
plt.plot(x1,y1)
plt.legend()

plt.show()
```

```
# GRAFICO DE DISPERSAO

x = [1,3,5,7,9]
y = [2,5,7,1,-1]
z = [100,200,280,40,80]

# título
plt.title("Grafico de Pontos")

# eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

# grafico de dispersao
plt.scatter(x,y, label = "Pontos", color = "g", marker = ".", s=z) 
plt.plot(x,y, color = "k", linestyle = "--")
plt.legend()

plt.show()
```
